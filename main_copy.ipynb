```python
temp = []
for i in range(128):
    b = bin(i)
    b = b[2:]
    l = len(b)
    if l<7:
        b = '0'*(7-l) + b
    b = list(b)
    b = list(map(int, b))
    temp.append(b)
```

```python
b = bin(10)
print(b, type(b))

def alph_to_bin(al):
    ch = bin(al)[2:]
    ch = list(ch)
    ch = list(map(int, ch))
    return ch

alph_to_bin(ord('a'))
```

```python
#Defining model architecture
model_dec = keras.Sequential([
    keras.layers.Dense(128, activation='relu', input_shape=(7,)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    # keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='softmax')
])
model_dec.summary()
```

```python
#Created to stop training when accuracy hits 100%
class CustomCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        if logs.get('accuracy') >= 1:
            self.model.stop_training = True
```

```python
# optimizer = keras.optimizers.Adam(lr=0.0001)
callback = CustomCallback()
model_dec.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#Creating the labels
labels = []
for i in range(128):
    l = [0] * 128
    l[i] = 1
    labels.append(l)
print(labels)
labels = tf.convert_to_tensor(labels)
labels.shape
```

```python
#Training the decryption modsel
epochs = 1000
model_dec.fit(results, labels, epochs=epochs, callbacks=[callback])
```

```python
#Saving the decryption model
model_dec.save('dec_model.keras')
```
